```python
import csv

# Open input CSV file
with open('transactions.csv', newline='') as input_file:
    reader = csv.DictReader(input_file)

    # Create output CSV file and write header
    with open('transaction_volume.csv', 'w', newline='') as output_file:
        writer = csv.writer(output_file)
        writer.writerow(['timestamp', 'volume'])

        # Iterate over each row in the input CSV file
        for row in reader:
            value = float(row['value']) / 10**18  # Convert value from wei to ether
            gas_price = float(row['gas_price']) / 10**9  # Convert gas price from Gwei to ether
            volume = value * gas_price  # Calculate transaction volume
            timestamp = int(row['block_timestamp'])  # Convert timestamp to integer
            writer.writerow([timestamp, volume])  # Write row to output CSV file
```

```python
import csv
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
# Open CSV file
with open('transaction_volume.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)

    # Read data into lists
    timestamps = []
    volumes = []
    for row in reader:
        timestamps.append(int(row['timestamp']))
        volumes.append(float(row['volume']))

    # Create plot
    plt.plot(timestamps, volumes)
    plt.xlabel('Timestamp')
    plt.ylabel('Volume')
    plt.title('Ethereum Transaction Volume')

    plt.show()
```
